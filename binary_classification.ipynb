```python
import os
import functools

import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
os.listdir('HRF_GLAUCOMA')
```

```python
glau_folders = [x for x in os.listdir() if x.endswith('GLAUCOMA')]
def get_paths(paths, folder):
    paths_to_files = [os.path.join(folder, x) for x in os.listdir(folder)]
    paths.extend(paths_to_files)   
    return paths
glau_paths = list(functools.reduce(get_paths, glau_folders, []))
```

```python
len(glau_paths)
```

```python
normal_folders = [x for x in os.listdir() if x.endswith('NORMAL')]
normal_paths = list(functools.reduce(get_paths, normal_folders, []))
```

```python
len(normal_paths)
```

```python
def load_image(path):
    img = tf.io.read_file(path)
    img = tf.image.decode_png(img)
    return img
```

```python
glau_folders
```

```python
i = 1
plt.figure(figsize=(20,12))
for i, folder in enumerate(glau_folders):
    path = [x for x in glau_paths if x.startswith(folder)][0]
    img = load_image(path)
    plt.subplot(1, len(glau_folders), i+1)
    plt.imshow(img.numpy())
    plt.axis('off')
    plt.title(folder)
#     for img in images:
#         plt.subplot(NUM_CLASSES, 5, i)
#         plt.imshow(img.numpy())
#         plt.axis('off')
#         plt.title(class_name)
#         i += 1
```
